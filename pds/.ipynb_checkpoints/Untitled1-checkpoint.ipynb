```python
import PyPDF2
import re
 
def csv_line_parter(number_of_columns, section) :
  parts_list = []
  section = section[:-1]
  for part in range(len(section) // number_of_columns) :
    line = ''
    for i in range(number_of_columns) :
      line = line + section[part*number_of_columns+i] + ','
#     print(line)
#     print(type(line))
    parts_list.append(line)
#     print(type(parts_list[0]))
  return parts_list
 
pdf_file = open('ObesityAnalysis.pdf','rb')
csv_file = open('ObesityAnalysis.csv', 'w')
 
page_content = ''
read_pdf = PyPDF2.PdfFileReader(pdf_file)
number_of_pages = read_pdf.getNumPages() // 4
 
for page_number in range(number_of_pages) :
  section1 = re.split("\n", read_pdf.getPage(page_number).extractText())
  section2 = re.split("\n", read_pdf.getPage(page_number + 160).extractText())
  section3 = re.split("\n", read_pdf.getPage(page_number + 320).extractText())
  section4 = re.split("\n", read_pdf.getPage(page_number + 480).extractText())
  part1 = csv_line_parter(7, section1)
  part2 = csv_line_parter(9, section2)
  part3 = csv_line_parter(8, section3)
  part4 = csv_line_parter(8, section4)
  for i, v in enumerate(part1) : page_content = page_content + ("{}{}{}{}\n".format(part1[i], part2[i], part3[i], part4[i])).replace(',\n', '\n')
 
csv_file.write(page_content)
```
