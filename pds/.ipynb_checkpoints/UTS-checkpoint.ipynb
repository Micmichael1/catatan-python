```python
import tabula.io as tab
import pandas as pd
# 'ObesityAnalysis.pdf'
```

```python
# Halaman tabel-tabel pada PDF
pages_patient = "1-160"
pages_diabetes = "161-320"
pages_risks = "321-480"
pages_age = "481-640"
```

```python
# Memproses data pasien (1 - 160)
dfs = tab.read_pdf('ObesityAnalysis.pdf', pages = pages_patient, stream = True, guess = False, columns = (100, 149, 280, 350, 422, 480), pandas_options = {'header' : None})
tb_patient = pd.DataFrame([]);
for df in dfs:
    tb_patient = pd.concat([tb_patient, df], ignore_index = True)
tb_patient = tb_patient.iloc[1:]
tb_patient.columns = ["PatientId", "Zone", "State", "District", "Age", "Gender", "Waist"]
tb_patient
```

```python
# Memproses data sejarah diabetes (161 - 320)
dfs = tab.read_pdf('ObesityAnalysis.pdf', pages = pages_diabetes, stream = True, guess = False, columns = (95, 150, 240, 290, 350, 400, 450, 500, 600), pandas_options = {'header' : None})
tb_diabetes = pd.DataFrame([]);
for df in dfs:
    tb_diabetes = pd.concat([tb_diabetes, df], ignore_index = True)
tb_diabetes = tb_diabetes.iloc[1:]
tb_diabetes.columns = ["Height", "Weight", "DiabetesSelfDelc", "Diafather", "Diamother", "Moderate", "Vigorous", "Dailyphysical", "Hba1c"]
tb_diabetes
```

```python
# Memproses data resiko diabetes (321 - 480)
dfs = tab.read_pdf('ObesityAnalysis.pdf', pages = pages_risks, stream = True, guess = False, columns = (110, 160, 250, 320, 370, 420, 470, 520), pandas_options = {'header' : None})
tb_risk = pd.DataFrame([]);
for df in dfs:
    tb_risk = pd.concat([tb_risk, df], ignore_index = True)
tb_risk = tb_risk.iloc[1:]
tb_risk.columns = ["DiabetesCalc", "Diabetes", "ClinicalNotes", "Bmi", "WcRiskScore", "BmiRiskScore", "BmiWcRiskScore", "IdrsWaist"]
tb_risk
```

```python
# Memproses data umur (481-640)
dfs = tab.read_pdf('ObesityAnalysis.pdf', pages = pages_age, stream = True, guess = False, columns = (100, 155, 200, 255, 310, 360, 440), pandas_options = {'header' : None})
tb_age = pd.DataFrame([]);
for df in dfs:
    tb_age = pd.concat([tb_age, df], ignore_index = True)
tb_age = tb_age.iloc[1:]
tb_age.columns = ["IdrsAge", "IdrsFamilyHistory", "IdrsBmi", "IdrsBmiWc", "IdrsPhysicalActivity", "Idrs", "IdrsWithBmi", "IdrsWithBmiWc"]
tb_age
```
